```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_df = pd.read_csv("output.csv")
raw_df
```

```python
raw_df.info()
```

```python
raw_df['created_at'] = pd.to_datetime(raw_df['created_at'])

raw_df['year'] = raw_df['created_at'].dt.year
raw_df['month'] = raw_df['created_at'].dt.month
raw_df['day'] = raw_df['created_at'].dt.day

raw_df
```

```python
raw_df = raw_df.drop(columns=['id', 'status'])
raw_df
```

```python
raw_df.isna().sum()
```

```python
raw_df['Commodity_Uom'].unique()
```

```python
conversion_factors = {          
    'Qui': 100,       
    'Kg': 1,
    '50 Kg': 50,
    '62 Kg': 62,
    '20 Kilogram': 20,
    '1 Kg': 1,
    'Gram': 0.001,
    'Gm': 0.001,
    'Per Gram': 0.001
}

raw_df['factor'] = raw_df['Commodity_Uom'].map(conversion_factors).fillna(1)     # fillna for nos 

raw_df['commodity_arrivals'] = raw_df['commodity_arrivals'] * raw_df['factor']
raw_df['commodity_traded'] = raw_df['commodity_traded'] * raw_df['factor']
raw_df['min_price'] = raw_df['min_price'] / raw_df['factor']
raw_df['modal_price'] = raw_df['modal_price'] / raw_df['factor']
raw_df['max_price'] = raw_df['max_price'] / raw_df['factor']

raw_df.loc[raw_df['Commodity_Uom'] != 'Nos', 'Commodity_Uom'] = 'Kg'

raw_df = raw_df.drop(columns=['factor'])

raw_df
```

```python
print(raw_df['Commodity_Uom'].value_counts())
```

```python
raw_df[(raw_df['commodity'] == 'LEMON') & (raw_df['Commodity_Uom'] == 'Kg')]
```

```python
nos_counts = raw_df[raw_df['Commodity_Uom'] == 'Nos']['commodity'].value_counts()

commodities_to_remove = nos_counts[nos_counts < 293].index

print("Before removing:", raw_df.shape)

raw_df = raw_df[~((raw_df['Commodity_Uom'] == 'Nos') & raw_df['commodity'].isin(commodities_to_remove))]

print("After removing:", raw_df.shape)
```

```python
raw_df[raw_df['Commodity_Uom'] == 'Nos']['commodity'].value_counts()
```

```python
both_units = (
    raw_df.groupby('commodity')['Commodity_Uom']
    .transform(lambda x: set(x) >= {"Nos", "Kg"})
)

df_both = raw_df[both_units]

result = df_both.groupby(['commodity', 'Commodity_Uom']).size().reset_index(name='count')

print(result)
```

```python
nos_to_kg = {
    "BETEL LEAVES": 0.0022,   
    "COCONUT": 2.0,           
    "COCONUT WITH HUSK": 2.1  
}

for commodity, factor in nos_to_kg.items():
    mask = (raw_df['commodity'] == commodity) & (raw_df['Commodity_Uom'] == 'Nos')
    raw_df.loc[mask, 'commodity_arrivals'] *= factor
    raw_df.loc[mask, 'commodity_traded'] *= factor
    raw_df.loc[mask, 'min_price'] /= factor
    raw_df.loc[mask, 'modal_price'] /= factor
    raw_df.loc[mask, 'max_price'] /= factor
    raw_df.loc[mask, 'Commodity_Uom'] = 'Kg'

print(raw_df[raw_df['commodity'].isin(nos_to_kg.keys())]['Commodity_Uom'].value_counts())
```
